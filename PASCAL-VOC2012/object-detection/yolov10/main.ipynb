```python
import sys
import cv2
import torch
import matplotlib.pyplot as plt
import supervision as sv
from ultralytics import YOLO
from torchvision.datasets import VOCSegmentation
from torch.utils.data import DataLoader
from torchvision.transforms import v2
import albumentations as A
from albumentations.pytorch.transforms import ToTensorV2

sys.path.append(r"../")

from dataset import PascalVOCDataset
```

```python
dataset = PascalVOCDataset(root=r"../../data", image_set="val", download=True, transform=A.Compose([
    A.Resize(height=160, width=256), 
    A.Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5), max_pixel_value=255), 
    ToTensorV2(transpose_mask=True), 
]))
    
model = YOLO("yolov10n.pt")
```

```python
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
for i, (image, _) in enumerate(dataloader):
    original_image = image[0].permute(1, 2, 0).numpy()
    results = model(image)[0]
    result_image = results.plot()
    f, axarr = plt.subplots(1,2)
    axarr[0].imshow(original_image)
    axarr[1].imshow(result_image)
    plt.show()
    
    if i == 50:
        break
```
